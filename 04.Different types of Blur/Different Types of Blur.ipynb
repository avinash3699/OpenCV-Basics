```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/elon.jfif")
```

```python
cv2_imshow(img)
```

## Gaussian Blur

```python
# Kernal Size
ksize = (9,9)

sigmaX = 0
```

```python
gaussianBlurImg = cv2.GaussianBlur(img, ksize, sigmaX)
```

```python
cv2_imshow(gaussianBlurImg)
```

## Median Blur

```python
img = cv2.imread("/content/elon.jfif")
```

```python
ksize = 9
```

```python
medianBlurImg = cv2.medianBlur(img, ksize)
```

```python
cv2_imshow(medianBlurImg)
```

## Bilateral Blur

```python
img = cv2.imread("/content/elon.jfif")
```

```python
k = 9
sigmaColor = sigmaSpace = 75 
```

```python
bilateralBlurImg = cv2.bilateralFilter(img, k, sigmaColor, sigmaSpace)
```

```python
cv2_imshow(bilateralBlurImg)
```
