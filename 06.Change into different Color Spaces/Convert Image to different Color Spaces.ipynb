```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/Elon.jpg')
```

## Original Image

```python
cv2_imshow(img)
```

___

## to GRAYSCALE

```python
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
```

___


## to HSV ( Hue, Saturation, Value )

```python
hsvImage = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsvImage)
```

___

## to HLS ( Hue, Lightness, Saturation )




```python
hlsImage = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
cv2_imshow(hlsImage)
```
