```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/elon.jfif")
```

```python
cv2_imshow(img)
```

```python
fontFace = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
thickness = 3
org = (50, 50)
color = (255, 255, 255)
```

```python
imgWithText = cv2.putText(img, "Hi!! This is Elon Musk", org, fontFace, fontScale, color, thickness)
```

```python
cv2_imshow(imgWithText)
```
