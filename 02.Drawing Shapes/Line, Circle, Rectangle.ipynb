## Line

```python
img = cv2.imread("/content/image.jpg")
```

```python
startCoordinates = (50,20)
endCoordinates = (50, 90)
thickness = 9
color = (255, 255, 255)
```

```python
imgWithLine = cv2.line(img, startCoordinates, endCoordinates, color, thickness)
```

```python
cv2_imshow(imgWithLine)
```

## Circle

```python
img = cv2.imread("/content/image.jpg")
```

```python
centerCoordinates = (100,100)
radius = 39
thickness = 9
color = (255, 255, 255)
```

```python
imgWithCircle = cv2.circle(img, centerCoordinates, radius, color, thickness)
```

```python
cv2_imshow(imgWithCircle)
```

## Rectangle

```python
img = cv2.imread("/content/image.jpg")
```

```python
startCoordinates = (50,20)
endCoordinates = (300, 90)
thickness = 9
color = (255, 255, 255)
```

```python
imgWithRectangle = cv2.rectangle(img, startCoordinates, endCoordinates, color, thickness)
```

```python
cv2_imshow(imgWithRectangle)
```
